### Python脚本中的问题
查看您的Python脚本 `2-伪二倍体文件统计.py`，问题可能出现在这几个地方：

#### 1. 多等位基因位点处理

查看您的Python脚本 `2-伪二倍体文件统计.py` 第71-81行：


### 🔥 关键问题

**`freq_label`, `type_label`, `special_label` 是基于整个位点的总AC计算的，但却被用于每个单独的ALT记录！**

让我详细解释：

1. **频率计算**基于 `total_ac = sum(ac_list)` 和 `af = total_ac / an`
2. **但是每个ALT都单独写入一行**，每行使用相同的频率标签
3. **结果**：同一个ALT可能因为**不同的total_ac值**而被分配**不同的频率类别**

这就解释了为什么会有重复！

## 🔧 解决方案

我找到了**问题的根源**和**解决方案**：

### 问题原因
您的Python脚本在处理**多等位基因位点**时有逻辑错误：

1. **频率计算**基于整个位点的`total_ac`
2. **但每个ALT单独写入一行**时，使用了相同的频率标签
3. **如果脚本被多次运行或有其他逻辑问题**，可能导致同一ALT被重复记录

### 具体的修复方案

**修改Python脚本的第40-81行部分：**

```python
# 原有代码问题：频率分类应该基于每个ALT的个体频率，而不是总频率
# 修复方案：为每个ALT分别计算频率

for i, (alt, ac_val) in enumerate(zip(var.ALT, ac_list)):
    if ac_val > 0:
        # 为每个ALT单独计算频率
        af_individual = ac_val / an  # 个体等位基因频率
        maf_individual = af_individual if af_individual <= 0.5 else 1 - af_individual
        
        # 基于个体MAF进行频率分类
        if maf_individual >= 0.05:
            freq_label_individual = "Common"
        elif maf_individual >= 0.01:
            freq_label_individual = "LowFreq"
        elif maf_individual >= 0.001:
            freq_label_individual = "Rare"
        else:
            freq_label_individual = "UltraRare"
        
        # singleton/doubleton应该基于个体AC
        if ac_val == 1:
            special_label_individual = "Singleton"
        elif ac_val == 2:
            special_label_individual = "Doubleton"
        else:
            special_label_individual = ""
        
        var_writer.writerow([
            var.CHROM,
            var.POS,
            var.REF,
            alt,
            ac_val,
            base,
            freq_label_individual,  # 使用个体频率标签
            type_label,             # 类型可以保持不变
            special_label_individual # 使用个体special标签
        ])
```
