        af_individual = ac_val / an  # ‰∏™‰ΩìÁ≠â‰ΩçÂü∫Âõ†È¢ëÁéá
        maf_individual = af_individual if af_individual <= 0.5 else 1 - af_individual

        if ac_val == 1:
            special_label_individual = "Singleton"
        elif ac_val == 2:
            special_label_individual = "Doubleton"
        else:
            special_label_individual = ""
        
        var_writer.writerow([
            var.CHROM,
            var.POS,
            var.REF,
            alt,
            ac_val,
            base,
            freq_label_individual,  # ‰ΩøÁî®‰∏™‰ΩìÈ¢ëÁéáÊ†áÁ≠æ
            type_label,             # Á±ªÂûãÂèØ‰ª•‰øùÊåÅ‰∏çÂèò

````python
# ‰∏¥Êó∂Ëß£ÂÜ≥ÊñπÊ°àÔºöÂØπÂΩìÂâçÊï∞ÊçÆËøõË°åÂéªÈáçÂ§ÑÁêÜ

print("=== ‰∏¥Êó∂Ëß£ÂÜ≥ÊñπÊ°àÔºöÊï∞ÊçÆÂéªÈáç ===")

# ÊñπÊ°à1Ôºö‰øùÁïôACÂÄºËæÉÂ§ßÁöÑËÆ∞ÂΩïÔºàÂèØËÉΩÊõ¥ÂáÜÁ°ÆÔºâ
print("ÊñπÊ°à1Ôºö‰øùÁïôÊØè‰∏™(POS,REF,ALT)ÁªÑÂêà‰∏≠ACÂÄºÊúÄÂ§ßÁöÑËÆ∞ÂΩï")
df_Global_Common_dedup1 = df_Global_Common.loc[df_Global_Common.groupby(['POS', 'REF', 'ALT'])['AC'].idxmax()]
print(f"ÂéªÈáçÂêéË°åÊï∞: {len(df_Global_Common_dedup1)} (ÂéüÊù•: {len(df_Global_Common)})")

# ÊñπÊ°à2Ôºö‰øùÁïôACÂÄºËæÉÂ∞èÁöÑËÆ∞ÂΩïÔºàÊõ¥‰øùÂÆàÔºâ
print("\nÊñπÊ°à2Ôºö‰øùÁïôÊØè‰∏™(POS,REF,ALT)ÁªÑÂêà‰∏≠ACÂÄºÊúÄÂ∞èÁöÑËÆ∞ÂΩï")
df_Global_Common_dedup2 = df_Global_Common.loc[df_Global_Common.groupby(['POS', 'REF', 'ALT'])['AC'].idxmin()]
print(f"ÂéªÈáçÂêéË°åÊï∞: {len(df_Global_Common_dedup2)} (ÂéüÊù•: {len(df_Global_Common)})")

# ÊñπÊ°à3ÔºöÈöèÊú∫‰øùÁïô‰∏Ä‰∏™ËÆ∞ÂΩï
print("\nÊñπÊ°à3ÔºöÈöèÊú∫‰øùÁïôÊØè‰∏™(POS,REF,ALT)ÁªÑÂêà‰∏≠ÁöÑ‰∏Ä‰∏™ËÆ∞ÂΩï")
df_Global_Common_dedup3 = df_Global_Common.drop_duplicates(subset=['POS', 'REF', 'ALT'], keep='first')
print(f"ÂéªÈáçÂêéË°åÊï∞: {len(df_Global_Common_dedup3)} (ÂéüÊù•: {len(df_Global_Common)})")

# Êé®Ëçê‰ΩøÁî®ÊñπÊ°à1Ôºà‰øùÁïôACËæÉÂ§ßÁöÑÔºâÔºåÂõ†‰∏∫ÂèØËÉΩ‰ª£Ë°®Êõ¥ÂáÜÁ°ÆÁöÑËÆ°Êï∞
print(f"\n=== Êé®Ëçê‰ΩøÁî®ÊñπÊ°à1 ===")
print(f"‰ΩøÁî®ÂéªÈáçÂêéÁöÑÊï∞ÊçÆÔºåÂîØ‰∏ÄÂèòÂºÇÊï∞ = {len(df_Global_Common_dedup1)} = ÂéüÂßãË°åÊï∞")

# È™åËØÅÂéªÈáçÊïàÊûú
print(f"\nÈ™åËØÅÔºöÂéªÈáçÂêéÊòØÂê¶ËøòÊúâÈáçÂ§çÔºü")
print(f"ÂéªÈáçÂêéÈáçÂ§çÁöÑ(POS,REF,ALT)ÁªÑÂêàÊï∞: {df_Global_Common_dedup1.duplicated(subset=['POS', 'REF', 'ALT']).sum()}")

# Â¶ÇÊûúÈúÄË¶ÅÁªßÁª≠ÂàÜÊûêÔºåÂèØ‰ª•Áî®ÂéªÈáçÂêéÁöÑÊï∞ÊçÆ
print(f"\nÂ¶ÇÊûúË¶ÅÁªßÁª≠Èü¶ÊÅ©ÂõæÂàÜÊûêÔºåÂª∫ËÆÆ‰ΩøÁî®Ôºö")
print(f"df_Global_Common_clean = df_Global_Common_dedup1")
````

````python
# ËÆ©Êàë‰ª¨Ê£ÄÊü•‰∏Ä‰∏ãÊï∞ÊçÆÁöÑÂü∫Êú¨‰ø°ÊÅØ
print("df_EA_Common ÂΩ¢Áä∂:", df_EA_Common.shape)
print("df_Global_Common ÂΩ¢Áä∂:", df_Global_Common.shape)

# Ê£ÄÊü•POS,REF,ALTÁªÑÂêàÁöÑÂîØ‰∏ÄÊÄß
print("\ndf_EA_Common ‰∏≠ (POS,REF,ALT) ÂîØ‰∏ÄÁªÑÂêàÊï∞:", df_EA_Common[['POS', 'REF', 'ALT']].drop_duplicates().shape[0])
print("df_Global_Common ‰∏≠ (POS,REF,ALT) ÂîØ‰∏ÄÁªÑÂêàÊï∞:", df_Global_Common[['POS', 'REF', 'ALT']].drop_duplicates().shape[0])

# Ê£ÄÊü•ÊòØÂê¶ÊúâÈáçÂ§çÁöÑË°å
print("\ndf_EA_Common ÊòØÂê¶ÊúâÈáçÂ§çË°å:", df_EA_Common.duplicated().any())
print("df_Global_Common ÊòØÂê¶ÊúâÈáçÂ§çË°å:", df_Global_Common.duplicated().any())

# Â¶ÇÊûúÊúâÈáçÂ§çÔºåÁúãÁúãÈáçÂ§çÁöÑÊï∞Èáè
if df_EA_Common.duplicated().any():
    print("df_EA_Common ÈáçÂ§çË°åÊï∞:", df_EA_Common.duplicated().sum())
if df_Global_Common.duplicated().any():
    print("df_Global_Common ÈáçÂ§çË°åÊï∞:", df_Global_Common.duplicated().sum())
````

````python
# Áé∞Âú®ÔºåÊàëË¶ÅÊ†πÊçÆPOS,REF,ALT‰∏âÂàóÊù•ÂêàÂπ∂df_EA_CommonÂíådf_Global_Common
# ËøôÂèØ‰ª•ÈÄöËøápandasÁöÑmergeÂáΩÊï∞Êù•ÂÆûÁé∞
df_common_merged = pd.merge(df_EA_Common, df_Global_Common, on=['POS', 'REF', 'ALT'], suffixes=('_EA', '_Global'),how='outer')
df_common_merged
# Áé∞Âú®Ôºådf_common_mergedÂåÖÂê´‰∫ÜÊâÄÊúâÂú®‰∏ú‰∫öÂíåÂÖ®ÁêÉ‰∏≠È¢ëÁéá‰∏∫CommonÁöÑÂèòÂºÇ
# ÊàëÂ∏åÊúõ‰∏∫ÊØèÁßçÂèòÂºÇÔºàÊåâÁÖßPOS,REF,ALT‰∏âÂàóÔºâÂàÜÈÖç‰∏Ä‰∏™Áã¨ÁâπÁöÑÁºñÂè∑
# Â¶ÇÊûúÊüê‰∏ÄË°åÁöÑPOS,REF,ALT‰∏âÂàóÁöÑÁªÑÂêàÂú®‰πãÂâçÁöÑË°å‰∏≠Âá∫Áé∞ËøáÔºåÂàô‰∏∫ËØ•Ë°åÂàÜÈÖçÁõ∏ÂêåÁöÑÁºñÂè∑
# Âê¶ÂàôÔºå‰∏∫ËØ•Ë°åÂàÜÈÖç‰∏Ä‰∏™Êñ∞ÁöÑÁºñÂè∑
df_common_merged['Variant_ID'] = df_common_merged.groupby(['POS', 'REF', 'ALT']).ngroup()
df_common_merged
````

````python
# ÊàëÊ£ÄÊü•‰∏Ä‰∏ãdf_common_merged['variant_id']ÁöÑÂîØ‰∏ÄÂÄºÊï∞Èáè
unique_variant_ids = df_common_merged['Variant_ID'].nunique()
unique_variant_ids
#? ‰∏∫‰ªÄ‰πàËøôÈáåÁöÑÂîØ‰∏ÄÊï∞ÈáèÂèòÊàê‰∫Ü257622ËÄå‰∏çÊòØ258238Ôºü
# ÂèØËÉΩÊòØÂõ†‰∏∫Âú®ÂêàÂπ∂ËøáÁ®ã‰∏≠ÔºåÊúâ‰∫õÂèòÂºÇÂú®‰∏ú‰∫öÂíåÂÖ®ÁêÉ‰∏≠ÈÉΩÂ≠òÂú®Ôºå‰ΩÜÂÆÉ‰ª¨ÁöÑPOS,REF,ALTÁªÑÂêàÂú®Êüê‰∏ÄÁªÑ‰∏≠ÊòØÂîØ‰∏ÄÁöÑÔºåËÄåÂú®Âè¶‰∏ÄÁªÑ‰∏≠‰∏çÊòØ„ÄÇ
````

````python
#todo ‰∏ã‰∏ÄÊ≠•ÊàëË¶ÅÂà∂‰ΩúÈü¶ÊÅ©ÂõæÁöÑÂáÜÂ§áË°®Ê†ºÔºåÁ¨¨‰∏ÄÂàó‰∏∫EAÔºåÁ¨¨‰∫åÂàó‰∏∫GlobalÔºåÊØè‰∏ÄË°åÊòØVariant_ID
# ËøôÊ†∑ÊàëÂ∞±ÂèØ‰ª•Áî®venn2ÂáΩÊï∞Êù•Âà∂‰ΩúÈü¶ÊÅ©Âõæ‰∫Ü
df_common_merged_EA_Ve = df_common_merged.loc[df_common_merged['Source_EA']=='East_Asia', ['Variant_ID']]
df_common_merged_Global_Ve = df_common_merged.loc[df_common_merged['Source_Global']=='Global', ['Variant_ID']]
df_common_merged_EA_Ve,df_common_merged_Global_Ve
````

````python
# ÈáçÂëΩÂêçÔºåÂ∞ÜVariant_IDÂàóÊîπ‰∏∫EA_Variant_IDÂíåGlobal_Variant_ID
df_common_merged_EA_Ve.rename(columns={'Variant_ID': 'EA_Variant_ID'}, inplace=True)
df_common_merged_Global_Ve.rename(columns={'Variant_ID': 'Global_Variant_ID'}, inplace=True)
````

````python
# Áé∞Âú®Áõ¥Êé•Êää‰∏§‰∏™dfÂº∫Ë°åÊãºÊàê‰∏Ä‰∏™df_Venn
df_Venn = pd.concat([df_common_merged_EA_Ve, df_common_merged_Global_Ve], axis=1)
# ËæìÂá∫‰∏∫‰∏Ä‰∏™csvÊñá‰ª∂Ôºå‰øùÁïô0‰ΩçÂ∞èÊï∞
df_Venn.to_csv(f'{BASE_DOR}/df_Venn.csv', index=False, float_format='%.0f')
````
